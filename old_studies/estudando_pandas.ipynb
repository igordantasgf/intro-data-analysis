- **Matplotlib Plotting of EUA Population**

```python
quadro_eua = pd.read_csv(dataset_eua, sep=',', header=0)
```

```python
latitude = list(quadro_eua['lat'])
longitude = list(quadro_eua['lon'])
tam = list(quadro_eua['pop']/100000)
colors = ['#edeeef']
plt.scatter(longitude, latitude, s=tam, c=colors)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
plt.savefig('usa_graph.png')
```
